```python
from langchain_community.document_loaders import PDFLoader  # or other loaders
from langchain.text_splitter import CharacterTextSplitter
from langchain_openai import OpenAIEmbeddings, ChatOpenAI
from langchain_core.vectorstores import InMemoryVectorStore
from langchain_core.chains import RetrievalQAWithSourcesChain

# 1. Load documents
loader = PDFLoader("../guidance_docs/Kenya-ARV-Guidelines-2022-Final-1.pdf")  # Adjust the path to your PDF file
documents = loader.load()
```

```python

# 2. Split documents
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=0)
texts = text_splitter.split_documents(documents)

# 3. Create embeddings
embeddings = OpenAIEmbeddings(model="text-embedding-3-large")

# 4. Store embeddings in vector store
vector_store = InMemoryVectorStore(embeddings)
vector_store.add_documents(texts)

# 5. Create retriever
retriever = vector_store.as_retriever()

# 6. Initialize LLM
llm = ChatOpenAI(model="gpt-4o-mini")

# 7. Create RAG chain
rag_chain = RetrievalQAWithSourcesChain.from_chain_type(llm, retriever=retriever)

# 8. Query the chain
response = rag_chain.invoke({"question": "Your question here"})
print(response)
```
